#  Strategic Data Joins ( similar to SQL JOINs) 



```python
# creating a dictionary dfs
dfs = {
    'customers': customers,
    'orders': orders,
    'order_items': order_items,
    'order_payments': order_payments,
    'order_reviews': order_reviews,
    'products': products,
    'sellers': sellers,
    'product_cat_translation': product_cat_translation
}
```

```python
# To Create "Master Orders" table that contains comprehensive information about each order, linked to customers, products, and reviews.

print("\n--- Performing Strategic Data Joins ---")
```

```python
# starting and using 'orders' table as the base
# This ensures we retain all orders and link relevant information to them.
master_df = dfs['orders'].copy()
print(f"Starting with 'orders' table. Shape: {master_df.shape}")

master_df.head()
```

```python
# Join with Customers (Left Join: Keep all orders, add customer info)
# Each order has a customer_id, and each customer_id maps to a unique customer.
# Use 'customer_id' from orders to link to 'customer_id' in customers.
master_df = pd.merge(master_df, dfs['customers'], on='customer_id', how='left')
print(f"After joining with 'customers'. Shape: {master_df.shape}")
print("Columns added: customer_unique_id, customer_zip_code_prefix, customer_city, customer_state")

master_df.head()
```

```python
# Join with Order Items (Left Join: Keep all orders, add item details. An order can have multiple items)
# This will duplicate rows for orders with multiple items. This is expected and often desired for item-level analysis.
master_df = pd.merge(master_df, dfs['order_items'], on='order_id', how='left')
print(f"After joining with 'order_items'. Shape: {master_df.shape}. Note: Rows will increase due to multiple items per order.")
print("Columns added: order_item_id, product_id, seller_id, shipping_limit_date, price, freight_value")

master_df.head()
```

```python
# Join with Products (Left Join: Add product details to each item)
master_df = pd.merge(master_df, dfs['products'], on='product_id', how='left')
print(f"After joining with 'products'. Shape: {master_df.shape}")
print("Columns added: product_category_name, product_name_lenght, product_description_lenght, product_photos_qty, product_weight_g, product_length_cm, product_height_cm, product_width_cm")

master_df.head()
```

```python
# Join with Product Category Name Translation (Left Join: Translate Brazilian category names)
# CRITICAL step for human-readable product categories.
master_df = pd.merge(master_df, dfs['product_cat_translation'], on='product_category_name', how='left')
print(f"After joining with 'product_category_name_translation'. Shape: {master_df.shape}")
print("Column added: product_category_name_english")

master_df.head()
```

```python
# Handle cases where translation might be missing (e.g., if a category name doesn't have a translation)
# For now, we'll fill NaN with the original Portuguese name, so we don't lose information.
master_df['product_category_name_english'] = master_df['product_category_name_english'].fillna(master_df['product_category_name'])
print("Filled missing English category names with original Portuguese names.")

master_df.head()
```

```python
# Join with Order Reviews (Left Join: Add review details to each order)
# One review per order_id. If an order has multiple items, the review will be duplicated for each item row.
master_df = pd.merge(master_df, dfs['order_reviews'], on='order_id', how='left')
print(f"After joining with 'order_reviews'. Shape: {master_df.shape}")
print("Columns added: review_id, review_score, review_comment_title, review_comment_message, review_creation_date, review_answer_timestamp")

master_df.head()
```

```python
# Join with Order Payments (Left Join: Add payment details to each order)
# An order can have multiple payments (e.g., split payments, different methods).
# This will also duplicate rows if an order has multiple payment types.
master_df = pd.merge(master_df, dfs['order_payments'], on='order_id', how='left')
print(f"After joining with 'order_payments'. Shape: {master_df.shape}")
print("Columns added: payment_sequential, payment_type, payment_installments, payment_value")

master_df.head()
```

```python
# Join with Sellers (Left Join: Add seller details to each item row)
# Each item is sold by a seller.
master_df = pd.merge(master_df, dfs['sellers'], on='seller_id', how='left')
print(f"After joining with 'sellers'. Shape: {master_df.shape}")
print("Columns added: seller_zip_code_prefix, seller_city, seller_state")

master_df.head()
```

```python
print("\nCompletely joined Master DataFrame created. Final Head:")
master_df.head()
```

```python
# Final master DataFrame info with number of nulls(missing values)
print("Final Master DataFrame Info:")
master_df.info()
print(master_df.isnull().sum())
```

#  Data Cleaning (after all the Joins are completed ) 


## Converting all relevant date/timestamp columns to datetime objects.


```python
date_columns = [
    'order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',
    'order_delivered_customer_date', 'order_estimated_delivery_date',
    'shipping_limit_date', 'review_creation_date', 'review_answer_timestamp'
]
```

```python
for col in date_columns:
    if col in master_df.columns:
        master_df[col] = pd.to_datetime(master_df[col], errors='coerce') # 'coerce' turns unparseable dates into NaT
        print(f"Converted '{col}' to datetime.")
    else:
        print(f"Warning: Column '{col}' not found in master_df.")
```

##  Missing Value Handling 


```python
print("\nMissing values before handling:")
missing_initial = master_df.isnull().sum()
missing_initial = missing_initial[missing_initial > 0]
print(missing_initial.sort_values(ascending=False))
```

### Handling missing values for specific columns:
 a) `review_comment_message` and `review_comment_title`:
 These are **text fields, often legitimately blank**. **Fill with empty string for NLP later**.


```python
master_df['review_comment_message'] = master_df['review_comment_message'].fillna('')
master_df['review_comment_title'] = master_df['review_comment_title'].fillna('')
print("Filled NaNs in 'review_comment_message' and 'review_comment_title' with empty strings.")
```

b) `order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date`:
 These indicate order processing/delivery status. NaNs mean the action didn't happen or isn't recorded.
 For 'order_delivered_customer_date', missing values imply the order was not delivered yet or lost.
 For now, we'll keep them as NaT; later when calculating durations, NaT will propagate.
 If an order wasn't delivered, its customer_delivered_date will be NaT. This is important info.
 For analysis requiring completed deliveries, we'll filter.


c) Product dimensions/weight (`product_name_lenght`, `product_description_lenght`, etc.):
 These often indicate missing product information. Can fill with median/mean or 0 if appropriate(choice is on us and what we want to do).
 
 For now, let's fill with **0, implying "not specified" or "no length"**. This decision depends on how we'll use them.


```python
product_numeric_cols = [
    'product_name_lenght', 'product_description_lenght', 'product_photos_qty',
    'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm'
]
for col in product_numeric_cols:
    if col in master_df.columns:
        master_df[col] = master_df[col].fillna(0)
        print(f"Filled NaNs in '{col}' with 0.")
```

d) `price`, `freight_value`: **Critical for Monetary Analysis**. Missing values would be problematic.


```python
# lets check if there are any. If so, investigate. For now, assume they are mostly present.
if master_df['price'].isnull().any() or master_df['freight_value'].isnull().any():
    print("Missing values detected in 'price' or 'freight_value'. Remove them so that they do not effect our analysis")
    # For now, let's drop rows where price or freight_value is missing, as they are important(central) to analysis.
    initial_rows = master_df.shape[0]
    master_df.dropna(subset=['price', 'freight_value'], inplace=True)
    print(f"Dropped {initial_rows - master_df.shape[0]} rows with missing 'price' or 'freight_value'.")
```

e) `review_score`: **Essential for Satisfaction Analysis**. If missing, it means no review.

We can fill with a **neutral value (0 or mean/median)** or considering as "no review" and handle separately.


```python
# For now, let's fill with 0 to indicate no review submitted.
master_df['review_score'] = master_df['review_score'].fillna(0)  
print("Filled NaNs in 'review_score' with 0 (indicating no review).")
```

f) `payment_type`, `payment_installments`, `payment_value`, `payment_sequential`:

 These are tied to order payments. If missing, it could indicate an issue with the order itself.
 We'll drop rows where `payment_type` is missing as it's fundamental for **payment analysis**.

```python
if master_df['payment_type'].isnull().any():
    initial_rows = master_df.shape[0]
master_df = master_df.dropna(subset=['payment_type'])
print(f"Dropped {initial_rows - master_df.shape[0]} rows with missing 'payment_type'.")
```

##  Outlier Identification 


 Check for **zero or negative prices/freight values (should be positive)**

```python
zero_price_items = master_df[(master_df['price'] <= 0) | (master_df['freight_value'] < 0)]
if not zero_price_items.empty:
    print(f"Found {len(zero_price_items)} rows with zero/negative price or freight_value. Investigating/removing...")
    # Decision: Remove these rows as they are likely data entry errors or non-revenue generating.
    master_df = master_df[~((master_df['price'] <= 0) | (master_df['freight_value'] < 0))]
    print(f"Removed {len(zero_price_items)} rows with problematic price/freight values.")
```

Check **product_photos_qty for unusually high values (could be errors)**
